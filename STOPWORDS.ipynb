#                                    Counting and plotting stopwords as bar graph

```python
import nltk
from nltk.corpus import stopwords
from collections import Counter
import matplotlib.pyplot as plt
import random

# Download stopwords
nltk.download('stopwords')

# Function to remove stopwords from a string
def remove_stop_words(string):
    stop_words = set(stopwords.words('english'))
    words = string.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    new_string = ' '.join(filtered_words)
    return new_string

# Generate a random input sentence
input_string = "The quick brown fox jumps over the lazy dog"

# Remove stopwords from the input sentence
result = remove_stop_words(input_string)
print("Original string:", input_string)
print("Modified string:", result)

# Count the frequency of stopwords in the input sentence
stop_words = set(stopwords.words('english'))
words = input_string.split()
stopword_freq = Counter([word for word in words if word.lower() in stop_words])

# Count the frequency of non-stopwords in the input sentence
non_stopword_freq = Counter([word for word in words if word.lower() not in stop_words])

# Plotting the bar graph
plt.bar(stopword_freq.keys(), stopword_freq.values(), color='b', alpha=0.5, label='Stopwords')
plt.bar(non_stopword_freq.keys(), non_stopword_freq.values(), color='r', alpha=0.5, label='Non-Stopwords')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Frequency of Stopwords vs Non-Stopwords')
plt.legend()
plt.show()
```

# Create array of frequency of stopwords

```python
from collections import Counter
from nltk.corpus import stopwords

def remove_stop_words(string):
    stop_words = set(stopwords.words('english'))
    words = string.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    new_string = ' '.join(filtered_words)
    stopwords_removed = [word for word in words if word.lower() in stop_words]
    return new_string, Counter(stopwords_removed)

input_string = "This is an example sentence to remove stop words from."
result, stopwords_freq = remove_stop_words(input_string)

print("Original string:", input_string)
print("Modified string:", result)
print("Frequency of stopwords:", stopwords_freq)
```
